```python
import quasar
import aiem
```

```python
filenames = ['../data/aiem/bchl-a-8-stack/tc/%s/exciton.dat' 
             % (A) for A in range(1,8+1)]
aiem_monomer = aiem.AIEMMonomer.from_tc_exciton_files(
    filenames=filenames,
    N=2,
    connectivity='linear',
    )
```

```python
optimizer = quasar.BFGSOptimizer.from_options()
optimizer = quasar.PowellOptimizer.from_options()
optimizer = quasar.JacobiOptimizer.from_options(jacobi_level=1)
```

```python
backend = quasar.QuasarSimulatorBackend()
```

```python
mcvqe = aiem.AIEM.from_options(
    backend=backend,
    aiem_monomer=aiem_monomer,
    nstate=2,
    optimizer=optimizer,
    # nmeasurement=10000,
    )
mcvqe.compute_energy()
```

```python
print(mcvqe.vqe_circuit)
```

```python
backend2 = quasar.CirqSimulatorBackend()
```

```python
circuit2 = backend2.build_native_circuit(mcvqe.vqe_circuit)
```
